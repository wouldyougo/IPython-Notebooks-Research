#Рыночно-нейтральные стратегии

## Парный трейдинг

Описание перечня функций стратегии и сопутсвующих индикаторов.

###1. Фукции формирования спреда.
На входе цены двух коррелированных ценных бумаг А и В:
    
    Pa и Pb
    
    Варианты определения цены для вычисления спреда.
    Цены Pat и Pbt могут браться как цены последней сделки в момент t, или как лучшие цены заявок  в стакане.    

####Спред - как отношение цен двух коррелированных ценных бумаг:

    St = Qa*Pat/Qb*Pbt
    
####Спред - как разность цен двух коррелированных ценных бумаг:

    St = Qat*Pat - Qbt*Pbt

где

    Qa - количество ценных бумаг А
    Qb - количество ценных бумаг B   
    Pat - цена бумаг А    
    Pbt - цена бумаг B    
         в момент времени t
         
#####Продажа Сперда (-St) это:
    
    -St = -Qa*Pat + Qb*Pbt
    т.е. продажа A и покупка B
    
#####Покупка Сперда (+St) это:
    
    +St = +Qa*Pat - Qb*Pbt
    т.е. покупка A и продажа B   

####1.1 Денежно нейтральный Спред
На входе:    

    Известны соотношения:
    Ka/b = Pa/Pb;     Kb/a = Pb/Pa
    где Pa Pb средние на некотором интервале цены инструментов.

Требуется определить количество Qat Qbt спреда St
    
    St = Qat*Pat - Qbt*Pbt -> 0

#####Определение доли каждой из бумаг спреда:
    через соотношение Kb/a:
    { Xa + Xb = 1; Xa - Xb*Kb/a = 0; 

    { Xb = 1 - Xa; Xa = Xb*Kb/a;  
    Xa = (1 - Xa)*Kb/a; Xa = Kb/a - Xa*Kb/a; Xa + Xa*Kb/a = Kb/a; Xa(1 + Kb/a) = Kb/a;
    => 
    Xa = Kb/a/(1 + Kb/a);
    Xb = 1 - Kb/a/(1 + Kb/a)
    ------------------------------
        или через соотношение Ka/b:
        { Xa + Xb = 1; Xa*Ka/b - Xb = 0; 

        { Xa = 1 - Xb; Xa*Ka/b = Xb; 
        (1 - Xb)*Ka/b = Xb; Ka/b - Xb*Ka/b = Xb; Ka/b = Xb + Xb*Ka/b; Ka/b = Xb*(1 + Ka/b);
        = >
        Xb = Ka/b/(1 + Ka/b);
        Xa = 1 - Ka/b/(1 + Ka/b); 
        ---------------------------------
            Проверка
            Pa = 2
            Pb = 1
            Pa/Pb = Ka/b = 2/1 = 2

            Xa + Xb = 1; Xa*2 = Xb; 
            Xb = 2/(1 + 2);
            Xa = (1 + 2 - 2)/(1 + 2) = 1/3

#####Пересчет параметров спреда   
    
    Ps - суммарная цена спреда определяется в зависимости от выделенных на стратегию средств.
    Ps = Qat*Pat + Qbt*Pbt

    Количество Qat Qbt будет браться в соответсвующих долях Xa и Xb.

    Wa = Xa*100
    Wb = Xb*100
    где
    Wa и Wb – % вес инструментов в составе среда;

    Xa/1 = Qa × Pa / Ps
    Xb/1 = Qb × Pb / Ps
    
    Qa = Xa*Ps / Pa 
    Qb = Xb*Ps / Pb    
    

#####Пересчет параметров спреда Упрощенный вариант

    Qat = K
    Qbt = K*Ka/b = K*Pa/Pb
    
    или 
    Qat = K*Kb/a = K*Pb/Pa
    Qbt = K
    
    где 
    K - некоторый целый множитель лотов 

    тогда формула спреда на основе текущего отношения:
    
    St =      K*Pat - Ka/b*K*Pbt
    St = Kb/a*K*Pat -      K*Pbt        

####1.2 Бета-нейтральный Спред

    St = Qat*Pat - Qbt*Pbt

Требуется вычислить коэффициенты спреда Qat и Qbt, через Бета-коэффициенты $\beta_a$ и $\beta_b$.

Доля каждого актива в структуре портфеля должна рассчитываться исходя из Беты обоих активов на момент совершения сделки. Кроме того, поскольку значение Беты не постоянно, периодически необходимо корректировать структуру портфеля, для поддержания его в рыночно-нейтральном состоянии.

####1.2.0 Вычисление Доходности
Для вычисления Бета-коэффициент используются "доходности", почти тоже самое что и цены, т.е. теже цены, только отцентрированные и нормированные относительно "начальной" цены (почти тоже самое делают перед вычислением корреляции). 

На входе 

    Px  - цены рыночного индекса;
    Pyi - цены iой бумаги спреде;

Нужно найти 

    Yi — доходность ценной бумаги i;
    X  — доходность рыночного индекса;

####1.2.1 Вычисление Бета-коэффициента
На входе:

    Ya — доходность ценной бумаги A;
    Yb — доходность ценной бумаги B
    X  — доходность рыночного индекса;
   
Требуется найти коэффициенты наклона βa и βb линейной регрессии

$Ya = \alpha + \beta_a X$

$Yb = \alpha + \beta_b X$

    βa — коэффициент наклона (Бета-коэффициент) для A;
    βb — коэффициент наклона (Бета-коэффициент) для B;    

В общем случае:
$Yi = \alpha + \beta_i X$   
    
    Yi — доходность ценной бумаги i;
    X  — доходность рыночного индекса;

#####Вычисление Бета-коэффициентов.
Фактически все что нам и нужно - это посчитать корреляцию доходности бумаг А и B с доходностью рыночного индекса X.

    Необходимо вычислить:
$\mathbf{\beta} = \frac{\mathbf{cov}_{XY}}{\mathbf{var}_{X}}$
    
#####Выборочная Ковариация
$\mathrm{cov}(X_{(n)},Y_{(n)}) = \frac1n\sum_{t=1}^n \left(X_t-\overline{X}\right)\left(Y_t-\overline{Y}\right)$

$\mathrm{cov}(X_{(n)},Y_{(n)}) = \frac1n\sum_{t=1}^n X_tY_t-\left(\frac1n\sum_{t=1}^nX_t\right)\left(\frac1n\sum_{t=1}^nY_t\right)$

#####Выборочная Дисперсия
   $\mathrm{Var}(X_n) = \sigma^2_x = \frac{1}{n} \sum\limits_{i=1}^n \left(X_i - \bar{X} \right)^2$

   $\mathrm{Var}(X_n) = \frac{1}{n}\sum\limits_{i=1}^nX_i^2-\left(\frac{1}{n}\sum\limits_{i=1}^nX_i\right)^2$


#####Промежуточные вычисления

$\frac1n\sum_{t=1}^nY_t$

$\frac1n\sum_{t=1}^nX_t$

$\left(\frac{1}{n}\sum\limits_{i=1}^nX_i\right)^2$

$\frac{1}{n}\sum\limits_{i=1}^nX_i^2$

$\frac1n\sum_{t=1}^n X_tY_t$

####1.2.2 Вычисление коэффициенов спреда для Бета-нейтрального парного трейдинга

Известны Бета-коэффициенты:

$\beta_a$ и $\beta_b$
для каждой из бумаг спреда 
    
    St = Qat*Pat - Qbt*Pbt

необходимо вычислить параметры (коэффициенты) спреда Qat и Qbt:

    Qat - количество ценных бумаг А
    Qbt - количество ценных бумаг B   
    , взятые в соответсвующих Бета-нейтральных долях Xa и Xb.

#####Определение доли каждой из бумаг спреда:

    { Xa + Xb = 1; Xa*βa = Xb*βb; 

    { Xa = 1 - Xb; Xa*βa = Xb*βb; 
      (1 - Xb)*βa = Xb*βb; βa - Xb*βa = Xb*βb; βa = Xb*βb + Xb*βa; βa = Xb*(βb + βa)
    } => 

    Xb = βa/(βb + βa)
    Xa = 1 - βa/(βb + βa)   
    
    или наоборот:
    { Xb = 1 - Xa; Xa*βa = Xb*βb; 
      Xa*βa = (1 - Xa)*βb; Xa*βa = βb - Xa*βb; Xa*βa + Xa*βb= βb; Xa(βa + βb)= βb; 
    } => 

    Xa = βb/(βa + βb); 
    Xb = 1 - βb/(βa + βb); 
    

#####Пересчет параметров спреда
    Значения Бета коэффициентов в каждый момент времени могут иметь разное значение, что также отразится на долевых коэффициентах, поэтому для каждого момента времени все коэффициенты пересчитываются.
На входе:
    
    Xa, Xb - доли  какждого инструмента
    Ps - суммарная цена спреда определяется в зависимости от выделенных на стратегию средств.
    Ps = Qat*Pat + Qbt*Pbt

Количество Qat Qbt будет браться в соответсвующих долях Xa и Xb.

    Wa = Xa*100
    Wb = Xb*100
    Wa и Wb – вес в % инструментов в составе среда;
    
    Xa = Qa × Pa / Ps
    Xb = Qb × Pb / Ps
    
    Qa = Xa*Ps / Pa 
    Qb = Xb*Ps / Pb

####3.0.2 Вычисление оклонения спреда от среднего с учетом волатильности
На входе: 
    
    St - спред двух коррелированных ценных бумаг А,В
    DSt = St - xMA(S,n)
    где xMA - некоторая скользящая средняя (MA) с периодом n от спреда S

    vDSt = (St - xMA(S,n))/σ(St,n)t
    vDSt =             DSt/σ(St,n)t    
    где σt - стандартное отклонения (с периодом n) спреда S.
    
    В общем случае
    xDSt = (St - xMA(S,n))/vol(St)
    xDSt =             DSt/vol(St)
    где vol(St) - некоторый показатель волатильности или скорости спреда.

####3.03. Открытие и закрытие позиции

Продажа Сперда (-St) это:
    
    -St = -Qa*Pat + Qb*Pbt
    т.е. продажа A и покупка B
    
Покупка Сперда (+St) это:
    
    +St = +Qa*Pat - Qb*Pbt
    т.е. покупка A и продажа B

####3.1 Многоуровневый Маркет мейкинг Определение уровней котирования
    На исторический данных по спреду определяется первый уровень отклонения спреда от своего среднего, который достигается наиболее часто.
    На исторический данных по спреду определяется максимальное историческое отклонения спреда от своего среднего. Это максимальное отклонение разбивается на некоторое количество уровней, при достижении которых совершаются сделки в поддержку спреда.
    Количество уровней котирования можно взять как уровень максимального исторического отклонения спреда.

####3.2 Многоуровневый Маркет мейкинг

На входе:
    
    S - спред двух коррелированных ценных бумаг А,В
    xMA - некоторая скользящая средняя (MA) спреда S
    xDS - некоторое оклонения спреда от среднего

В стратегии Многоуровневый Маркет мейкинг задаем количество уровней, и задаем значения для каждого уровня,  на которых будут совершаться сделки покупки и продажи спреда.
    
    lev           - количество уровней в каждую сторону от нулевого
    L = lev*2 + 1 - всего уровней котирования с учетом нулевого
    
    lev = 3
    тогде  L = 7
    
    Сделка Продажа или Покупка Сперда осуществляетя при достижении очередного уровня котирования значением отклонения спреда от своего скользящего средного.
    Сделка Продажа или Покупка Сперда не должна осуществляетя при достижении отклонением очередного уровня, если предудущая позиция была открыта на этом же уровне.

####3.2.1 Дополнительные индикаторы и фильтры

#####Направление средней

$\mathrm{Trend}_{SMA(A/B)_t}=\left({\left({\frac{\mathrm{SMA}_{(A/B)_t}}{\mathrm{SMA}_{(A/B)_{t-1}}}}\right)-1}\right)\cdot100$

    Интерпретация:
    "Ещё одним важным фактором является относительная сила и направление средней скользящей. 
    Если её величина находиться на значительном удалении от нуля и совпадает с направлением Абсолютного отклонения, это говорит о возможном схлопывании отклонения, но не за счет возврата соотношения к средней, а за счет приближения самой средней к текущему соотношению. Если же направление средней скользящей противоположно текущему Абсолютному отклонению, то очень высока вероятность возврата соотношения к своему среднему значению в ближайшем будущем".

#####Ширина исторического канала

$\mathrm{Width}_{(A/B)}=\left({\frac{\mathrm{Ratio}_{(A/B)}\mathrm{max}}{\mathrm{Ratio}_{(A/B)}\mathrm{min}}}\right)\cdot100$

    процентное отношение исторического максимума к историческому минимуму.

####3.2.2 Вспомогательные алгоритмы

##### Выход по стоп-лос
Если намечается выход спреда за исторические границы диапазона можно предусмотреть закрытие всей или части позиции по стоп-лосу, для последующего перестроения самого спреда.

##### Пробойная Трендследящая стратегия
В паре к работаюещей стратегии по алгоритму Многоуровневый Маркет мейкинг, можно запустить Пробойная Трендследящую стратегию, которая будет открывать позиции по тому же спреду при выходе спреда за исторические границы диапазона. *Фактически здесь ловим черных лебедей.*

##### Трендследящая стратегия
В паре к работаюещей стратегии по алгоритму Многоуровневый Маркет мейкинг, можно запустить простую Трендследящую стратегию, которая будет открывать позиции по тому же спреду при появлении тренда также и внутри исторического диапазона. Т.е. при сильном и явном изменении скользящей средней спреда.
Этот вариант позволит компенсировать потери стратегии по алгоритму Многоуровневый Маркет мейкинг при появлении длительного тренда по спреду.
