##Парный трейдинг (частный случай арбитража и баскет трейдинга)
###Общие моменты для стратегий парного трейдинга

Парный трейдинг - "Рыночно-нейтральная" инвестиционная стратегия, основанная совершении логически связанных сделок, направленных на извлечение прибыли из разницы в ценах 2-х коррелированных ценных бумаг. "Рыночно-нейтральная" стратегия означает, что доходность стратегии не зависит от общего направления движения рынка.

Стратегия Парного трейдинг основана на извлечение прибыли при кратковременном изменении соотношения стоимости 2-х коррелированных ценных бумаг, т.е. на изменении стоимости Спреда.

Спред - (в нашем случае для парного трейдинга) разность цен двух коррелированных ценных бумаг.
Спред - (в случае баскет трейдинга) разность цен двух коррелированных пакетов (корзин) ценных бумаг.
Приведенный в этом файле алгоритм нужно будет, обобщить до алгоритма баскет трейдинга.

###1. Вычисление спреда.
На входе цены двух коррелированных ценных бумаг А и В:
    
    Pa и Pb

####Спред - как отношение цен двух коррелированных ценных бумаг:

    St = Qa*Pat/Qb*Pbt
    
Часто говорят про торговлю отношением, насколько я понял, Спред как отношение используется только для принятия решения, фактически при совершении сделки на покупку или продажу спреда имеем разность от покупки и продажи бумаг.

####Спред - как разность цен двух коррелированных ценных бумаг:

    St = Qa*Pat - Qb*Pbt

где

    Qa - количество ценных бумаг А
    Qb - количество ценных бумаг B   
    Pat - цена бумаг А    
    Pbt - цена бумаг B    
         в момент времени t
         
Продажа Сперда (-St) это:
    
    -St = -Qa*Pat + Qb*Pbt
    т.е. продажа A и покупка B
    
Покупка Сперда (+St) это:
    
    +St = +Qa*Pat - Qb*Pbt
    т.е. покупка A и продажа B
    
Варианты определения цены для вычисления спреда.
Цены Pat и Pbt могут браться как цены последней сделки в момент t, или как лучшие цены заявок в стакане.

В общем случае значения Qa Qb также могут зависеть от в момента времени t (напремер с течение времени может меняться среднее отношение A/B из за фундаментальных факторов).

    St = Qat*Pat - Qbt*Pbt   

Изменение значения Спреда S с течение времени t не зависит от изменения (движения) рынка (или зависит в значтельно меньшей степени).
   
    Пример 1.1
    Средние значения цен
    A = 100
    B = 90
    при торговле 1 лотом каждой из бумаг, т.е. Qat = Qbt = 1    
    S = 100 - 90 = 10
    Спред практически не зависит от движения рынка

    Пример 1.2
    Средние значения цен    
    A = 1000
    B = 90
    при торговле 1 лотом
    S = 1000 - 90 = 910
    Спред фактически зависит от движения рынка на 90%. Такой спред мало пригоден и фактически не отличается от направленной торговли одним инструментом.
    
Как раз от выбора Qat, Qbt и зависит насколько Спред независим от движения рынка. Далее рассмотрим способы вычисления Qat и Qbt.

####1.1 Денежно нейтральный портфель
Часто стремятся чтобы сред был около нуля.
    
    St = Qat*Pat - Qbt*Pbt = 0
    
Как раз для этого случая вычисляют отношение Pat/Pbt.

    Пример 1.1.1
    Средние значения цен
    A = 100
    B = 90
    A/B = 100/90 = 1.11 = Ka/b

    тогда 
        Qat = 1
        Qbt = Ka/b = A/B
        или 
        Qat = Kb/a = B/A
        Qbt = 1
        
        тогда формула спреда на основе текущего отношения
            St =    1*Pat - Ka/b*Pbt
            St = Kb/a*Pat -    1*Pbt            

    St = 1*Pat - 1.1*Pbt

    Торговли дробными лотами нет, поэтому если хотят максимально уравновесит открытую позицию по деньгам, могут применить множитель для количества лотов например кратный 10.
    
    Qat = 1*10
    Qbt = A/B*10 = 1.1*10 = 11
    
    St = 10*Pat - 11*Pbt
    S = 10*100 - 11*90 = 1000 - 990 = 10
    
    Спред фактически не зависит от движения рынка. При это абсолютное значение спреда осталось прежним = 10 как и в примере 1, но открываема позиция стала в 10 раз более денежно нейтральная по отношению к рынку, чем при торговле 1 лотом каждой из бумаг.
    
    Пример 1.1.2
    Средние значения цен    
    A = 1000
    B = 90
    A/B = 1000/90 = 11.11 = Ka/b

            St =    1*Pat - Ka/b*Pbt
            St = Kb/a*Pat -    1*Pbt            
    тогда
    St = 1*Pat - 11.11*Pbt
    S  = 1*1000 - 11*90 = 1000 - 990 = 10    
    
    или 
    St = 0,09*Pat - 1*Pbt    
    
    с множителем 10 (вариант получается менее точным чем предыдущий)
    St = 0,9*Pat - 10*Pbt    
    S  = 1*1000 - 10*90 = 1000 - 900 = 100
    
    или с множителем 100
    St = 9*Pat - 100*Pbt  
    S  = 9*1000 - 100*90 = 9000 - 9000 = 0

Денежно нейтральный спред хорошо подходит для фьючерсов на евро и доллар.

    Пример 1.1.3
    На входе
        Фьючерс на доллар Si       [S/R] = [Доллар США/Рубль]
        Фьючерс на евро   Eu       [Е/R] = [Евро/Рубль]
        Фьючерс на евро/доллар Ed  [Е/S] = [Евро/Доллар США]
 	 	
    Отношение Eu/Si ~> Ed
    
            St =  Qat*Pat -  Qbt*Pbt
            St =    1*Pat - Ka/b*Pbt
            St = Kb/a*Pat -    1*Pbt            
    Pat = Eu
    Pbt = Si
    Ked = Ed (~ Eu/Si)
    Ka/b = Ked = Ed
    Kb/a = 1/Ked = 1/Ed    
    
    Тогда спред евро - доллар:
        S = 1*Eu - Ka/b*Si                
        
        S = 1*Eu - Ked*Si                        
        S = 1*Eu - Ked*Si = [Е/R] - [Е/S]*[S/R] = [Е/R]- [Е/R]

        S = (1/Ked)*Eu - 1*Si
        S = (1/Ked)*Eu - 1*Si = (1/[Е/S])*[Е/R] - [S/R] = [S/R]- [S/R]

Денежно нейтральный спред не всегда рыночно нейтральный.

####1.2 Бета-нейтральный портфель
Денежно нейтральный спред не всегда рыночно нейтральный. Т.е. две бумаги могут быть сильно коррелированы, иметь в определенный момент времены близкие абсолютные значения цен, но при этом быть сильно разными по волатильности (по скорости изменения цен), т.е. сильно отличаться по реакции на внешние факторы.
В этом случае для построения спреда применяют Бета-нейтральный подход.

    St = Qat*Pat - Qbt*Pbt

Требуется вычислить параметры (коэффициенты) спреда Qat и Qbt, через Бета-коэффициенты $\beta_a$ и $\beta_b$.

Для этого доходность каждой из частей спреда сравнивают с доходностью рынка (или базового портфеля).    

Доля каждого актива в структуре портфеля должна рассчитываться исходя из Беты обоих активов на момент совершения сделки. Кроме того, поскольку значение Беты не постоянно, периодически необходимо корректировать структуру портфеля, для поддержания его в рыночно-нейтральном состоянии.

####1.2.1 Бета-коэффициент
Бета-коэффициент - является коэффициентом $\beta$ в линейной регрессии Y относительно X

   $Y = \alpha + \beta X$

Для стратегии парного трейдинга Нам нужно вычислить 2 Бета-коэффициента - для каждой составляющей спреда, для бумаги А и B - соответсвующие коэффициенты наклона βa и βb. Фактически все что нам и нужно - это посчитать корреляцию бумаг А и B с рыночным индексом X.

В нашем случае интересуют коэффициенты наклона (Бета-коэффициент) $\beta_a$ и $\beta_b$:
   
   $A = \alpha + \beta_a X$
   
   $B = \alpha + \beta_b X$   

Для вычисления Бета-коэффициент используются "доходности", почти тоже самое что и цены, т.е. теже цены, только отцентрированные и нормированные относительно "начальной" цены (почти тоже самое делают перед вычислением корреляции).

#####Основные сведения про Бета-коэффициент:    
    Бета-коэффициент  — показатель, рассчитываемый для ценной бумаги или портфеля ценных бумаг. Является мерой рыночного риска, отражая изменчивость доходности ценной бумаги (портфеля) по отношению к доходности другого портфеля, в роли которого часто выступает среднерыночный портфель (рыночный индекс).

    Согласно рыночной модели, предложенной Уильямом Шарпом, доходность отдельной ценной бумаги, можно описать уравнением:

   $r_i=\alpha_{iI}+\beta_{iI}r_I+\varepsilon_{iI}$

    Где:
    ri — доходность ценной бумаги;
    rI — доходность рыночного индекса;
    βiI — коэффициент наклона (Бета-коэффициент);
    αiI — коэффициент смещения (Альфа-коэффициент);
    εiI — случайная погрешность. 

Бета-коэффициент можно вычислить через МНК или еще проще через корреляцию.

Все что нужно для вычисления:
$\mathbf{\beta} = \frac{\mathbf{cov}_{XY}}{\mathbf{var}_{X}}$
    
   $\mathrm{Cov}$ — ковариация оцениваемой и эталонной величины, 
   $\mathrm{Var}$ — дисперсия эталонной величины.

####1.2.2 Вычисление коэффициенов спреда для Бета-нейтрального парного трейдинга
Теперь, когда у нас известны Бета-коэффициенты $\beta_a$ и $\beta_b$ для каждой из бумаг спреда, необходимо вычислить параметры (коэффициенты) спреда Qat и Qbt.
    
    St = Qat*Pat - Qbt*Pbt
    где

    Qat - количество ценных бумаг А
    Qbt - количество ценных бумаг B   
    , взятые в соответсвующих Бета-нейтральных долях Xa и Xb.

Зная Бета-коэффициенты βa и βb, можем определить долю X каждой из бумаг спреда в Бета-нейтрально портфеле:

\begin{cases}\mathrm{X}_\mathrm{A}+\mathrm{X}_\mathrm{B}=1\\(\mathrm{X}_\mathrm{A}\cdot{\beta_a})=(\mathrm{X}_\mathrm{B}\cdot{\beta_b})\end{cases}

    Xa = 1 - Xb
    Xa*βa = Xb*βb
    
    (1 - Xb)*βa = Xb*βb
    βa - Xb*βa = Xb*βb
    βa = Xb*βb + Xb*βa
    βa = Xb*(βb + βa)
    
    Xb = βa/(βb + βa)
    Xa = 1 - βa/(βb + βa)
    как-то так, если нигде не ошибся )

Значения Бета коэффициентов в каждый момент времени могут иметь разное значение, что также отразится на долевых коэффициентах, поэтому для каждого момента времени все коэффициенты пересчитываются.

Фактическое количество Qat Qbt будет браться в зависимости от выделенных на стратегию средств в соответсвующих долях Xa и Xb.

    Пример 1.1.2 (из википедии): 
    За 2009 год Бета-коэффициент акций «Лукойл» составил 1,01. 
    В случае же с акциями «Газпром нефть» Бета-коэффициент составил 0,87.

\begin{cases}\mathrm{X}_\mathrm{LKOH}+\mathrm{X}_\mathrm{SIBN}=1\\(\mathrm{X}_\mathrm{LKOH}\cdot{1,01})=(\mathrm{X}_\mathrm{SIBN}\cdot{0,87})\end{cases}
 
     тогда доля X каждой из бумаг в спреде:

\begin{cases}\mathrm{X}_\mathrm{LKOH}=0,46
\quad\mathrm{X}_\mathrm{SIBN}=0,54\end{cases}

###2. Вычисление отклонения
    Стратегии Парного трейдинга основаны на совершении сделок по спреду при отлонении его от среднего значения. При увеличении спреда совершается сдела по его продаже, при уменьшении спреда совершается сдела по его покупке, в ожидании на возврат их соотношения в ближайшем будущем к своему среднему значению. Когда же соотношения стоимости активов достигнет своего среднего значения, совершается обратная сделка с целью фиксации прибыли.
Следовательно нам нужно вычислить оклонение спреда от среднего (в общем случае от скользящего среднего).

####2.1 Вычисление оклонения спреда от среднего.
На входе спред двух коррелированных ценных бумаг А,В:
    
    St

На небольших временных интервалах ключевым фактором, на основе которого принимается решение о том, что один актив недооценен или переоценен относительно другого, является отклонение текущего соотношения от значения его скользящей средней.
    
    DSt = St - xMA(S,n)
    где xMA - некоторая скользящая средняя (MA) с периодом n от спреда S

т.е. на основе раницы St - xMA(S,n) принимается решение о покупке или продаже спреда.

Для Спреда - как отношения:
Если текущее отклонение соотношения от его средней скользящей находится глубоко в отрицательной зоне, это означает, что актив, расположенный в числителе, недооценен относительно актива, расположенного в знаменателе. И наоборот, если текущее отклонение соотношения от его средней скользящей находится глубоко в положительной зоне, это означает, что актив, расположенный в числителе, переоценен относительно актива, расположенного в знаменателе.

####2.2 Вычисление оклонения спреда от среднего с учетом волатильности
На входе: 
    
    St - спред двух коррелированных ценных бумаг А,В
    DSt = St - xMA(S,n)
    где xMA - некоторая скользящая средняя (MA) с периодом n от спреда S

Абсолютное отклонение спреда от среднего не учитывает волатильности Соотношения. "Абсолютное отклонение может показать благоприятный момент для входа в позицию, поскольку оно находиться на максимальных или минимальных исторических значениях, образующих уровни поддержки и сопротивления, однако если в текущий момент у Соотношения будет повышенная волатильность, то вполне вероятно, что эти уровни могут быть пробиты. Учесть риск повышенной волатильности позволяет отклонение соотношения, выраженное в величинах Стандартного отклонения σ."

    vDSt = (St - xMA(S,n))/σ(St,n)t
    vDSt =             DSt/σ(St,n)t    
    где σt - стандартное отклонения (с периодом n) спреда S.
    
"отклонение, выраженное через σ, показывает меньшую величину, чем Абсолютное отклонение, при повышенной волатильности, и большую величину, чем Абсолютное отклонение, при пониженной волатильности. Это означает, что в периоды относительно стабильного соотношения стоимости активов, входить в позицию можно при небольшой величине Абсолютного отклонения, поскольку, несмотря на меньшую прибыль с одной сделки, соотношение чаще возвращается к среднему значению, обеспечивая большее количество возможных сделок и повышая тем самым доходность. И наоборот, в периоды нестабильности соотношения, величина отклонения становиться больше, а его возврат к средней происходит реже, в связи с чем сделок рекомендуется совершать меньше и на большем удалении от средней."

    В общем случае
    xDSt = (St - xMA(S,n))/vol(St)
    xDSt =             DSt/vol(St)
    где vol(St) - некоторый показатель волатильности или скорости спреда.

###3. Стратегии торговли, открытие и закрытие позиции
    Стратегии Парного трейдинга заключаются в совершении сделок по спреду при отлонении его от среднего значения. При увеличении спреда совершается сдела по его продаже, при уменьшении спреда совершается сдела по его покупке, в ожидании на возврат их соотношения в ближайшем будущем к своему среднему значению. Когда же соотношения стоимости активов достигнет своего среднего значения, совершается обратная сделка с целью фиксации прибыли.

Продажа Сперда (-St) это:
    
    -St = -Qa*Pat + Qb*Pbt
    т.е. продажа A и покупка B
    
Покупка Сперда (+St) это:
    
    +St = +Qa*Pat - Qb*Pbt
    т.е. покупка A и продажа B

#####В Парном трейдинге можно существенно снизить общий риск портфеля путем одновременной торговли по нескольким парам. ####3.1 Маркет мейкинг

Стратегия Маркет мейкинга заключаются в совершении сделок по спреду при отлонении его от среднего значения на определенный уровень.

Т.е. чем чаще спред отклонится от своего среднего и вернется к нему, тем больше сделок мы совершим, отлавливая эти колебания.

####3.2.2 Вспомогательные алгоритмы

##### Выход по стоп-лос
Если намечается выход спреда за исторические границы диапазона можно предусмотреть закрытие всей или части позиции по стоп-лосу, для последующего перестроения самого спреда.

#####Разбить пару
И вариант требующий понимания ситуации с инструментами в паре, это разбить пару, закрыв полностью или частично одну из ног и возможно увеличив количество лотов на другой стороне.
Для того что бы делать осознанный выбор давайте вернёмся к основам парного трейдинга и посмотрим откуда берётся прибыль или убытки по паре.

##### Пробойная Трендследящая стратегия
В паре к работаюещей стратегии по алгоритму Многоуровневый Маркет мейкинг, можно запустить Пробойная Трендследящую стратегию, которая будет открывать позиции по тому же спреду при выходе спреда за исторические границы диапазона. *Фактически здесь ловим черных лебедей.*

##### Трендследящая стратегия
В паре к работаюещей стратегии по алгоритму Многоуровневый Маркет мейкинг, можно запустить простую Трендследящую стратегию, которая будет открывать позиции по тому же спреду при появлении тренда также и внутри исторического диапазона. Т.е. при сильном и явном изменении скользящей средней спреда.
Этот вариант позволит компенсировать потери стратегии по алгоритму Многоуровневый Маркет мейкинг при появлении длительного тренда по спреду.
